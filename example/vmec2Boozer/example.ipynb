### This is an example of how to use the python package `tfpy` to transform a quantity from the vmec coordinates to the Boozer coordinates.

```python
import numpy as np
import matplotlib.pyplot as plt 
from tfpy.vmec import VMECOut
from tfpy.boozXform import BoozerForm
import matplotlib
matplotlib.rcParams['text.usetex'] = True
```

```python
case_name = "QAS"
vmec_file = "../datas/VMECDatas/wout_"+case_name+".nc" 
vmeclib = VMECOut(vmec_file)
boozer_file = "../datas/BoozerDatas/boozer_wout_"+case_name+".nc"
boozerlib = BoozerForm(boozer_file)
```

```python
b_v2boozer = vmeclib.transBoozer(-1, vmeclib.getB(-1))
```

```python
fig, ax = plt.subplots()
boozerlib.getB(-1).plot_plt(fig=fig, ax=ax, fill=True, onePeriod=True)
b_v2boozer.plot_plt(fig=fig, ax=ax, fill=False, onePeriod=False)
```

#### Also, you can transform multiple quantities at the same time.

```python
_surf, _lam = vmeclib.getSurface(-1)
_omgea = vmeclib.getOmega(-1)
rzomega_v2boozer = vmeclib.transBoozer(-1, [_surf.r, _surf.z, _omgea])

boozerSurf = boozerlib.getSurface(-1)
```

```python
fig, ax = plt.subplots()
boozerSurf.r.plot_plt(fig=fig, ax=ax, fill=True, onePeriod=True)
rzomega_v2boozer[0].plot_plt(fig=fig, ax=ax, fill=False, onePeriod=False)
```

```python
fig, ax = plt.subplots()
boozerSurf.z.plot_plt(fig=fig, ax=ax, fill=True, onePeriod=True)
rzomega_v2boozer[1].plot_plt(fig=fig, ax=ax, fill=False, onePeriod=False)
```

```python
fig, ax = plt.subplots()
boozerSurf.omega.plot_plt(fig=fig, ax=ax, fill=True, onePeriod=True)
rzomega_v2boozer[2].plot_plt(fig=fig, ax=ax, fill=False, onePeriod=False)
```
